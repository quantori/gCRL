```python

# imports
import pickle
import numpy as np
import pandas as pd
import anndata as ad
import scanpy as sc
import random
from pathlib import Path
import sys
repo_parent = Path('./code/SERGIO').resolve().parent
sys.path.insert(0, str(repo_parent))
from SERGIO.sergio import sergio

# parameters
grn_folder = '../generated_data/'
noise_params = 1
decays=0.8 
sampling_state=15
noise_type='dpd'
```

```python
sim = sergio(number_genes=282, # to be read from the grn_folder
             number_bins = 1, # always one
             number_sc = 100, # to be taken as a input
             noise_params = noise_params, 
             decays=decays, 
             sampling_state=sampling_state, 
             noise_type=noise_type)
```

```python
sim.build_graph(input_file_taregts = grn_folder + 'input_targets.txt', 
                input_file_regs = grn_folder + 'input_regs.txt', 
                shared_coop_state = 2)
```

```python
sim.simulate()
expr = sim.getExpressions()
expr_clean = np.concatenate(expr, axis = 1)

df = pd.DataFrame(expr_clean)
df
```
