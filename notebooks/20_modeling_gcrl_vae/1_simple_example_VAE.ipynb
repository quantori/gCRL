## gCRL-VAE applied on a simple simulated dataset

- Load example adata (with TF communities already specified)
- Compute eigengenes
- Train gCRL-VAE

```python
# ensuring all packages are reloaded each time I run a cell
%load_ext autoreload
%autoreload 2
```

```python
# importing
import scanpy as sc
from gcrl.training.train_gcrl_vae import train_gcrl_vae
from gcrl.training.train_gcrl_vae import VAEConfig
from gcrl.grn.eigengenes import compute_eigengenes
```

```python
# loading data
adata = sc.read_h5ad("../../data/example/adata.h5ad")
adata
```

```python
adata.X
```

```python
adata.obs
```

```python
# computing eigengenes
compute_eigengenes(adata)
adata
```

```python
# gCRL-VAE configuration
cfg = VAEConfig()
cfg.batch_size = 128
cfg.outdir = "../../results/simulated/example/vae"
cfg
```

```python
# training gCRL-AE
res = train_gcrl_vae(adata=adata, cfg = cfg)
```
